```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import glob
from PIL import Image
import matplotlib.animation as animation
#%matplotlib notebook
```

```python
gif_name = 'HydrophilicSideChain1[101]'

data_name = 'HydrophilicSideChain1/nafion_out_HydrophilicSideChain1_{}.csv'
data_num = np.arange(0,101,10)
data_num = np.append(data_num, 101)
data_num
data = [[],[],[],[],[],[],[],[],[],[],[],[]]
for i,j in enumerate(data_num):
    data[i] = pd.read_csv(data_name.format(j),index_col=0,names=['X','Y','Z'])
```

```python
for i in range(80):
    list = np.where((data[0].iloc[:,2]>i-20)&(data[0].iloc[:,2]<i-19))
    data_2D = data[11].iloc[list]
    data_2D = data_2D.iloc[:,:2]

    a_2D=data_2D.loc['A']
    b_2D=data_2D.loc['B']
    c_2D=data_2D.loc['C']
    w_2D=data_2D.loc['W']

 #   %matplotlib notebook
    #データをarray型に
    x1,y1 = np.array(a_2D["X"]),np.array(a_2D["Y"])
    x2,y2 = np.array(b_2D["X"]),np.array(b_2D["Y"])
    x3,y3 = np.array(c_2D["X"]),np.array(c_2D["Y"])
    x4,y4 = np.array(w_2D["X"]),np.array(w_2D["Y"])

    # グラフ作成
    plt.ioff()
    fig = plt.figure(figsize=(6,6))
    plt.xlim(-20, 60)
    plt.ylim(-20, 60)

    plt.scatter(x1, y1)
    plt.scatter(x2, y2)
    plt.scatter(x3, y3)
    plt.scatter(x4, y4)

    fig.savefig("cross_section_pic/img_{}.png".format(i))

print('acomplished')
```

```python
#フォルダ名を入れます 
folderName = "cross_section_pic" 
 
#該当フォルダから画像のリストを取得。読み込みたいファイル形式を指定。ここではpng 
picList = glob.glob(folderName + "/*.png")
     
#figオブジェクトの作成
fig = plt.figure(figsize=(6,6))
 
#figオブジェクトから目盛り線などを消す
plt.tick_params(bottom=False,
                left=False,
                right=False,
                top=False)
plt.tick_params(labelbottom=False,
                labelleft=False,
                labelright=False,
                labeltop=False)
 
#空のリスト作成
ims = []
     
#画像ファイルを空のリストの中に1枚ずつ読み込み
for i in range(len(picList)):
         
    #読み込んで付け加えていく
    tmp = Image.open(picList[i])
    ims.append([plt.imshow(tmp)])
     
#アニメーション作成 
ani = animation.ArtistAnimation(fig, ims, interval=100, repeat=False)

ani.save("cross_section_gif/{}.gif".format(gif_name))
```

```python
%matplotlib nbagg

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig = plt.figure()
x = np.arange(0, 10, 0.1)

ims = []
for a in range(50):
    y = np.sin(x - a)
    im = plt.plot(x, y, "r")
    ims.append(im)

ani = animation.ArtistAnimation(fig, ims)
plt.show()
```
