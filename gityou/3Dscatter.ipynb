```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
data_name = 'nafion_out_HydrophilicSideChain1_{}.csv'
data_num = np.arange(0,101,10)
data_num = np.append(data_num, 101)
data_num
data = [[],[],[],[],[],[],[],[],[],[],[],[]]
for i,j in enumerate(data_num):
    data[i] = pd.read_csv(data_name.format(j),index_col=0,names=['X','Y','Z'])
```

```python
data[11]
```

```python
a = data[0].loc['A']
a
```

```python
b = data[0].loc['B']
b
```

```python
c = data[0].loc['C']
```

```python
w = data[0].loc['W']
w
```

```python
%matplotlib notebook
#データをarray型に
x1,y1,z1 = np.array(a["X"]),np.array(a["Y"]),np.array(a["Z"])
x2,y2,z2 = np.array(b["X"]),np.array(b["Y"]),np.array(b["Z"])
x3,y3,z3 = np.array(c["X"]),np.array(c["Y"]),np.array(c["Z"])
x4,y4,z4 = np.array(w["X"]),np.array(w["Y"]),np.array(w["Z"])

# グラフ作成
fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(x1, y1, z1,s=0.1)
ax.scatter(x2, y2, z2,s=0.1)
ax.scatter(x3, y3, z3,s=0.1)
ax.scatter(x4, y4, z4,s=0.1)
# 軸ラベルの設定
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.show()
```
