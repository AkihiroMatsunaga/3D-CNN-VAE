```python
def meshgrid3D(xnum,ynum,znum):
    x = np.arange(xnum).reshape(-1,1,1)
    y = np.arange(ynum).reshape(-1,1)
    z = np.arange(znum).reshape(1,-1)
    x3d = np.tile(x,(1,xnum,xnum))
    y3d = np.tile(y,(ynum,1,ynum))
    z3d = np.tile(z,(znum,znum,1))
    return x3d.flatten(),y3d.flatten(),z3d.flatten()
```

```python
x,y,z = meshgrid3D(80,80,80)
xy = np.concatenate([x.reshape(-1,1),y.reshape(-1,1)],1)
xyz = np.concatenate([xy,z.reshape(-1,1)],1)

xyzd0 = np.concatenate([xyz,data[0].flatten().reshape(-1,1)],1)
xyzd1 = np.concatenate([xyz,data[1].flatten().reshape(-1,1)],1)
xyzd2 = np.concatenate([xyz,data[2].flatten().reshape(-1,1)],1)
xyzd3 = np.concatenate([xyz,data[3].flatten().reshape(-1,1)],1)
```

```python
def plot_scatter(plot_data,kind,cmap):
    fig = plt.figure()
    ax1 = Axes3D(fig)#fig.add_subplot(411 , projection='3d')
    #sc = plt.scatter3D(x, y, z, s=1, c=data[0])
    scat = ax1.scatter(plot_data['X'], plot_data['Y'], plot_data['Z'], "o", c=plot_data['data'], s=10, cmap=cmap, lw=1, edgecolors='pink')
    #plt.colorbar(ax1)
    pp = fig.colorbar(scat, ax=ax1, orientation="vertical")
    pp.mappable.set_clim(np.min(data[kind]),np.max(data[kind]))

    plt.show()
```
